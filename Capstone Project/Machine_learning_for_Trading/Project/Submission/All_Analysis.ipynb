```python
# Writing Utility functions

import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, HTML
%matplotlib inline  


def symbol_to_path(symbol, base_dir="data"):
    """Return CSV file path given ticker symbol."""
    return os.path.join(base_dir, "{}.csv".format(str(symbol)))


def get_data(symbols, dates):
    """Read stock data (adjusted close) for given symbols from CSV files."""
    df = pd.DataFrame(index=dates)
    if 'SNP_500' not in symbols:  # add SPY for reference, if absent
        symbols.insert(0, 'SNP_500')

    for symbol in symbols:
        # TODO: Read and join data for each symbol
        df_temp=pd.read_csv("Data/{}.csv".format(symbol), index_col='Date',parse_dates=True,usecols=['Date','Adj Close']
                           ,na_values=['nan'])
        df_temp = df_temp.rename(columns={'Adj Close': symbol})
        df = df.join(df_temp) #use default how='left' 
        if symbol == 'SNP_500':
            df = df.dropna(subset=["SNP_500"])
    return df

    
def plot_data(df, title="Stock prices", xlabel="Date", ylabel="Price"):
    """Plot stock prices with a custom title and meaningful axis labels."""
    ax = df.plot(title=title, fontsize=12,figsize = (15,10))
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    plt.show()


    
def plot_selected(df, columns, start_index, end_index, title="Stock prices"):
    """Plot the desired columns over index values in the given range."""
    df_new = df.ix[start_index : end_index,columns]
    ax = df_new.plot(title=title,fontsize=10,figsize = (15,10))
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    plt.show()
    
def normalize_data(df):
    """Normalize data by uisng the first row of the dataframe"""
    return df/df.ix[0,:]

def get_rolling_mean(values, window):
    """Return rolling mean given values, using specified window size"""
    return pd.rolling_mean(values, window=window)

def get_rolling_std(values, window):
    """Return rolling standard deviation of given values, using specified window size"""
    return pd.rolling_std(values, window=window)

def get_bollinger_bands(rm, rstd):
    """Return upper lower bollinger bands."""
    upper_band = rm + rstd * 2
    lower_band = rm - rstd * 2
    return upper_band, lower_band

def compute_daily_return(df):
    daily_returns = (df / df.shift(1)) - 1
    daily_returns.ix[0, :] = 0
    return daily_returns[1:]

def cummulative_return(df):
    cummulative_return = (df.ix[-1] / df.ix[0]) - 1
    return cummulative_return
    
    
def test_run():
    # Define a date range
    dates = pd.date_range('2011-01-01', '2016-12-01')

    # Choose stock symbols to read
    symbols = ['AAPL', 'GOOG', 'AMZN', 'FACEBOOK', 'GE', 'MICROSOFT','GLD']
    # Get stock data
    df = get_data(symbols, dates)
    df = df.sort_index(ascending=True, axis=0)
    
      # Fill empty trade dates with forward filling dates first and then backward filling
    df.fillna(method="ffill", inplace="True")
    df.fillna(method="bfill", inplace="True")
    
    
    plot_data(df[['SNP_500','FACEBOOK']], title="Stock prices")
    
    #slice by row and column
    df_new = df.ix['2015-01-01':'2016-12-01',['AAPL', 'GOOG', 'AMZN', 'FACEBOOK', 'GE', 'GLD','MICROSOFT','SNP_500']]
   
    
    #normalize and plot the data all the stocks
    
    plot_data(normalize_data(df), title="Normalized stock prices")
    
    #normalize and plot few selected stocks
    plot_selected(normalize_data(df),['GE','SNP_500'], '2011-01-01', '2016-12-01',title="Normalized stock prices")
    
    
    #Compute Global Statistics

    
    #Compute Bollinger Bands
    #1. Compute rolling mean
    rm_SNP_500 = get_rolling_mean(df_new['GE'],window=20)
    
    #2. Compute rolling std
    rstd_SNP_500 = get_rolling_std(df_new['GE'],window=20)
    
    #3. Compute upper and lower bollinger bands
    upper_band, lower_band = get_bollinger_bands(rm_SNP_500, rstd_SNP_500)
    
    #plot bollinger bands
    ax = df_new['GE'].plot(title="GE bollinger bands", label='GE',fontsize=10,figsize = (15,10))
    rm_SNP_500.plot(label='Rolling Mean', ax=ax ,fontsize=5)
    upper_band.plot(label='Upper Band', ax=ax ,fontsize=5)
    lower_band.plot(label='Lower Band', ax=ax ,fontsize=5)
    
    #Add axis labels and legends
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    ax.legend(loc='upper left')
    plt.figure(figsize=(40,10))
    #plt.show()
    
    #Compute and plot daily return for all stocks
    daily_returns = compute_daily_return(df.ix['2015-01-01':'2015-02-01',['AAPL', 'GOOG', 'AMZN', 'FACEBOOK', 'GE','GLD' ,'MICROSOFT','SNP_500']])
    plot_data(daily_returns, title="Daily returns", ylabel="Daily returns")
    
    #Compute and plot daily return for selected stocks compared to SNP_500
    
    daily_returns = compute_daily_return(df.ix['2015-01-01':'2015-02-01',['GE', 'SNP_500']])
    plot_data(daily_returns, title="Daily return selected few", ylabel="Daily returns")
    
    print "---- Printing commulative returns for all the stocks for the period between '2015-01-01' and '2016-12-01----"
    print "\n"

    
    
    for i in symbols:
        cret = cummulative_return(df_new[i])
        print "commulative return for "+ i +" is:"
        per = cret*100
        per = "%.2f" % per
        print str(per)+"%"
    
   
    #Compute and plot daily return for SNP_500
    daily_returns = compute_daily_return(df[['SNP_500']])
    plot_data(daily_returns, title="Daily return for SNP_500", ylabel="Daily returns")  
    
    #Compute and plot daily return for SNP_500
    daily_returns_GE = compute_daily_return(df[['GE']])
    plot_data(daily_returns, title="Daily return for GE", ylabel="Daily returns")  
    
    
    #compute Mean, Stndard Deviation and Kurtosis for SNP_500
    print "Mean, Stndard Deviation and Kurtosis for SNP_500"
    mean = daily_returns.mean()
    print "Mean =", mean
    std = daily_returns.std()
    print "std =", std
    kur = daily_returns.kurtosis()
    print "kurtosis =", kur
    
    #compute Mean, Stndard Deviation and Kurtosis for GE
    print "Mean, Stndard Deviation and Kurtosis for GE"
    mean = daily_returns_GE.mean()
    print "Mean =", mean
    std = daily_returns_GE.std()
    print "std =", std
    kur = daily_returns_GE.kurtosis()
    print "kurtosis =", kur
    
    
    #Plot histogram for multiple stocks on the same chart - GE and SNP_500
    
    dr_hist = compute_daily_return(df[['SNP_500', 'GE']])
    dr_hist['SNP_500'].hist(bins=30 , label ="SNP_500",)
    dr_hist['GE'].hist(bins=30 , label ="GE")
    plt.legend(loc='upper right')
    GE_mean = dr_hist['GE'].mean()
    SNP_mean = dr_hist['SNP_500'].mean()
    GE_std = dr_hist['GE'].std()
    SNP_std = dr_hist['SNP_500'].std()
    plt.axvline(GE_mean,color='w',linestyle='dashed',linewidth=2)
    plt.axvline(SNP_mean,color='r',linestyle='dashed',linewidth=2)
    plt.axvline(GE_std,color='w',linestyle='dashed',linewidth=2)
    plt.axvline(-GE_std,color='w',linestyle='dashed',linewidth=2)
    plt.axvline(SNP_std,color='r',linestyle='dashed',linewidth=2)
    plt.axvline(-SNP_std,color='r',linestyle='dashed',linewidth=2)
    plt.show()
    
    
    #Scatter plots and co-relation
    dr_scatter = compute_daily_return(df)
    
    #Scatter plot SNP_500 vs GE
    dr_scatter.plot(kind='scatter',x='SNP_500',y='GE')
    
    #Fitting line with beta and alpha for GE
    beta_GE, alpha_GE = np.polyfit(dr_scatter['SNP_500'],dr_scatter['GE'],1)
    plt.plot(daily_returns['SNP_500'],beta_GE*daily_returns['SNP_500'] + alpha_GE, '-',color='r')
    print "beta_GE =", beta_GE
    print "alpha_GE =", alpha_GE
    
    
    #Scatter plot SNP_500 vs GLD
    dr_scatter.plot(kind='scatter',x='SNP_500',y='GLD')
      #Fitting line with beta and alpha for GLD
    beta_GLD, alpha_GLD = np.polyfit(dr_scatter['SNP_500'],dr_scatter['GLD'],1)
    plt.plot(daily_returns['SNP_500'],beta_GLD*daily_returns['SNP_500'] + alpha_GLD, '-',color='r')
    print "beta_GLD =", beta_GLD
    print "alpha_GLD =", alpha_GLD
    
    #Calculate Co-orelation coefficient
    print dr_scatter.corr(method='pearson')
    
    
    #### RUN PORTFOLIO STATS
    
    
    # Define a date range
    dates = pd.date_range('2014-12-01', '2016-12-01')

    # Choose stock symbols to read
    symbols = ['AAPL','GOOG', 'AMZN', 'FACEBOOK', 'GE', 'MICROSOFT']
    
    # Get stock data
    df = get_data(symbols, dates)
    
    # Fill empty trade dates with forward filling dates first and then backward filling
    df.fillna(method="ffill", inplace="True")
    df.fillna(method="bfill", inplace="True")

    df_port = df.drop('SNP_500', axis=1)
    
    # Normalize stock prices
    df_normalized = normalize_data(df_port)
    print "\n"
    print "normalized: "
    print df_normalized.head()
    print "\n"
    
    # Portfolio allocation for each stock
    allocation = [0.166, 0.166, 0.166, 0.166, 0.166, 0.166]
    df_allocation = df_normalized * allocation
    print "allocated: "
    print df_allocation.head()
    print "\n"
    
    # starting value for each stock -- dividing 1 million equally among all stocks
   
    starting_value = [1000000]
    df_values = df_allocation * starting_value
    print "Portfolio Stock value : "
    print display(df_values.head())
    print "\n"
    
    # Portfolio value by day
    portfolio_values = pd.DataFrame()
    portfolio_values['portfolio_value'] = df_values.sum(axis=1)
    print "Portfolio values: "
    print display(portfolio_values.head())
    print "\n"
    
    # Compute daily returns on the portfolio
    daily_returns_portfolio = compute_daily_return(portfolio_values)
    print "Portfolio Daily returns: "
    print display(daily_returns_portfolio.head())
    print "\n"
    
    
    ### Plotting Portfolios statistics
    
    print "Following is the plotted portfolio statistics"
    print "\n"
    plot_data(portfolio_values, title="Portfolio daily values starting with 10000000", xlabel="Date", ylabel="Price")
    print "Follwoing is the portfolio value normalzed plot"
    print "\n"
    plot_data(normalize_data(portfolio_values), title="Portfolio value mormalized", xlabel="Date", ylabel="Price")
    
    print "Following is the portfolio value normalized plot vs normalized SNP_500"
    print "\n"
    normalize_data(portfolio_values).plot(title="my portfolio vs. SNP_500 (normalized)",fontsize=12,figsize=(12,10), label="my portfolio")
    normalize_data(df)['SNP_500'].plot(title="my portfolio vs. SPY (normalized)",fontsize=12,figsize=(12,10), label="SNP_500")
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()
    
    
    
    print "Fund allocations by symbol: symbols = {}, allocation = {}".format([x for x in symbols if x != 'SNP_500'], allocation)
    print "\n"
    
    # Cumulative return
    cr = cummulative_return(portfolio_values)
    print "Cumulative return: ", cr
    
    # Average daily return
    print "Average daily return: ", daily_returns_portfolio.mean()
    

    # standard deviation 
    print "Daily standard deviation: ", daily_returns_portfolio.std()
    

    # Sharpe ratio
    Sharpe_ratio = np.sqrt(252) * (daily_returns_portfolio.mean())/daily_returns_portfolio.std()
    print "Sharpe ratio: ", Sharpe_ratio
    print "\n"
      
    
if __name__ == "__main__":
    test_run()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize as spo
import numpy as np
%matplotlib inline  


def symbol_to_path(symbol, base_dir="data"):
    """Return CSV file path given ticker symbol."""
    return os.path.join(base_dir, "{}.csv".format(str(symbol)))


def get_data(symbols, dates):
    """Read stock data (adjusted close) for given symbols from CSV files."""
    df = pd.DataFrame(index=dates)
    if 'SNP_500' not in symbols:  # add SPY for reference, if absent
        symbols.insert(0, 'SNP_500')

    for symbol in symbols:
        # TODO: Read and join data for each symbol
        df_temp=pd.read_csv("Data/{}.csv".format(symbol), index_col='Date',parse_dates=True,usecols=['Date',
                                                                                                     'Adj Close']
                           ,na_values=['nan'])
        df_temp = df_temp.rename(columns={'Adj Close': symbol})
        df = df.join(df_temp) #use default how='left' 
        if symbol == 'SNP_500':
            df = df.dropna(subset=["SNP_500"])
    return df
  
    
def normalize_data(df):
    """Normalize data by uisng the first row of the dataframe"""
    return df/df.ix[0,:]

def compute_daily_return(df):
    daily_returns = (df / df.shift(1)) - 1
    daily_returns.ix[0, :] = 0
    return daily_returns[1:]

def cummulative_return(df):
    cummulative_return = (df.ix[-1] / df.ix[0]) - 1
    return cummulative_return


def plot_data(df, title="Stock prices", xlabel="Date", ylabel="Price"):
    ax = df.plot(title=title, figsize=(12,10), fontsize=12)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    plt.show()
 

def find_alloc(df, alloc_func, num_symbols):
    
    # guess initial allocation 
    allocgss = [0.166, 0.166, 0.166, 0.166, 0.166, 0.166]
    
    # has to bound the allocations betwen 0 and 1. setting the min, max pair for each element in x
    bounds = [(0,1.0) for i in range(num_symbols)]
    
    # call optimizer to minimize alloc_function

    result = spo.minimize(alloc_func, allocgss, args=(df,), bounds=bounds, method='SLSQP', 
                           options={'disp':True}, 
                            constraints=({ 'type': 'eq', 'fun': lambda inputs: 1.0 - np.sum(inputs)})).x
    
    return result
    
    
def alloc_function(alloc, df):
    
    trading_period_days = 252
    starting_value = 1000000
    
    #Normalize stock and apply allocation
    df_normalized = normalize_data(df)
    df_normalized_allocation = df_normalized * alloc 
    
    # Calculate portfolio value by day
    starting_value = [1000000]
    # normalize portfolio
    df_normed_alloc_value = df_normalized_allocation * starting_value 
    portfolio_values = df_normed_alloc_value.sum(axis=1)
    
    
    # Portfolio value by day
    portfolio_val = pd.DataFrame()
    portfolio_val['portfolio_value'] = portfolio_values
    
    # Compute daily returns on the portfolio
    daily_returns_portfolio = compute_daily_return(portfolio_val)
   
    
    # Higher the sharp ratio the better (Reward / Risk) , so we should minimize the negative sharp ratio
    sharpe_ratio = -1 * np.sqrt(trading_period_days) * (daily_returns_portfolio.mean()/daily_returns_portfolio.std())
    return sharpe_ratio


def run():
    
    dates = pd.date_range('2014-12-01', '2016-12-01') # Define a date range
    symbols = ['AAPL','GOOG', 'AMZN', 'FACEBOOK', 'GE', 'MICROSOFT'] # Choose stock symbols to read
    df = get_data(symbols, dates) # Get stock data
    df = df.sort_index(ascending=True, axis=0)
    print "List of stocks"
    print symbols
    print "\n"
    
    df_port = df.drop('SNP_500', axis=1) # Drop the SNP_500 index
    df_port.fillna(method="ffill", inplace="True") # Forward fill empty trade dates
    df_port.fillna(method="bfill", inplace="True") # backfill empty trade dates 
    
    num_symbols = len([x for x in symbols if x != 'SNP_500'])
     #try to get all optimal allocations
    optimal_alloc = find_alloc(df_port, alloc_function, num_symbols)
    
    
    print "optimal allocations:"
    print ("AAPL: {0:.3f}".format(optimal_alloc[0]))
    print ("GOOG: {0:.3f}".format(optimal_alloc[1]))
    print ("AMZN: {0:.3f}".format(optimal_alloc[2]))
    print ("FACEBOOK: {0:.3f}".format(optimal_alloc[3]))
    print ("GE: {0:.3f}".format(optimal_alloc[4]))
    print ("MICROSOFT: {0:.3f}".format(optimal_alloc[5]))
    
    
    #### Plotting optimal Portfolio statistics
    
    
    # Normalize stock prices
    df_normalized = normalize_data(df_port)
    print "\n"
    
   
    
    # Portfolio allocation for each stock
    allocation = []
    for i in range(len(optimal_alloc)):
        allocation.append(round(optimal_alloc[i],3))
    print allocation 
    print "\n"
    
    
    df_allocation = df_normalized * allocation
    
    
    # starting value for each stock -- dividing 1 million equally among all stocks
   
    starting_value = [1000000]
    df_values = df_allocation * starting_value
    
    # Portfolio value by day
    portfolio_values = pd.DataFrame()
    portfolio_values['portfolio_value'] = df_values.sum(axis=1)
    
    # Compute daily returns on the portfolio
    daily_returns_portfolio = compute_daily_return(portfolio_values)
  
    
    print "\n"
    print "Following is the plotted optimal portfolio statistics:"
    print "\n"
    plot_data(portfolio_values, title="Optimzed Portfolio daily values starting with 10000000", xlabel="Date", ylabel="Price")
    print "Following is the optimal portfolio value normalized plot"
    print "\n"
    plot_data(normalize_data(portfolio_values), title="Optimzed Portfolio value normalized", xlabel="Date", ylabel="Price")
    
    print "Following is the optimal portfolio value normalized plot vs normalized SNP_500"
    print "\n"
    normalize_data(portfolio_values).plot(title="my optimized portfolio vs. SNP_500 (normalized)",fontsize=12,figsize=(12,10), label="my portfolio")
    normalize_data(df)['SNP_500'].plot(title="my optimized portfolio vs. SPY (normalized)",fontsize=12,figsize=(12,10), label="SNP_500")
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()
    
    
    # Cumulative return
    print "Fund allocations by symbol: symbols = {}, allocation = {}".format([x for x in symbols if x != 'SNP_500'], allocation)
    print "\n"
    cr = cummulative_return(portfolio_values)
    print "Cumulative return: ", cr
    
    # Average daily return
    print "Average daily return: ", daily_returns_portfolio.mean()

    # standard deviation 
    print "Daily standard deviation: ", daily_returns_portfolio.std()
    
    # Sharpe ratio
    
    Sharpe_ratio = np.sqrt(252) * (daily_returns_portfolio.mean())/daily_returns_portfolio.std()
    print "Sharpe ratio: ", Sharpe_ratio
    print "\n"
    
    
    
    print "Scaller plot between my optimized portfolio and SNP_500"
    
    daily_return_all = compute_daily_return(df)
    dr_scatter = pd.concat([daily_return_all, daily_returns_portfolio], axis=1)
    
    dr_scatter.plot(kind='scatter',x='SNP_500',y='portfolio_value')
    
     #Fitting line with beta and alpha for Portfolio and printring Beta and alpha
     
    beta_portfolio, alpha_portfolio = np.polyfit(dr_scatter['SNP_500'],dr_scatter['portfolio_value'],1)
    plt.plot(dr_scatter['SNP_500'],beta_portfolio*dr_scatter['SNP_500'] + alpha_portfolio, '-',color='r')
    print "beta_AAPL = " +str(beta_portfolio)
    print "alpha_AAPL = " +str(alpha_portfolio)
    
    print "\n"
    if float(alpha_portfolio) > 0.00:
        print "aplha value is positive, so in an average my optimized portfolio is performing little bit better than the SNP_500"
    elif float(alpha_portfolio) < 0.00:
        print "alpha is negative, so in an average my optimized portfolio is returning little bit less compared to the market"
    else:
        pass
    
    print "\n"
    
    #Calculate Co-orelation coefficient
    print "Co-relation between my optimzed portfolio and the market (SNP_500):"
    print "\n"
    print dr_scatter[['SNP_500', 'portfolio_value']].corr(method='pearson')
    

    print "\n"
    
    
     #Plot histogram for my portfolio and campare with SNP_500
        
    #Compute mean and std for my portfolio
    mean = dr_scatter['portfolio_value'].mean()
    print "Mean for my optimized portfolio=", mean
    
    mean_SNP = dr_scatter['SNP_500'].mean()
    
    std = dr_scatter['portfolio_value'].std()
    print "standard deviation for my portfolio=", std
    
    std_SNP = dr_scatter['SNP_500'].std()
    
    
    #Compute Kurtosis
    kur = dr_scatter['portfolio_value'].kurtosis()
    print "kurtosis for my optimized portfolio =", kur
    
    dr_scatter['SNP_500'].hist(bins=20 , label ="SNP_500")
    dr_scatter['portfolio_value'].hist(bins=20 , label ="portfolio_value")
    plt.legend(loc='upper right')
    
    
    #draw std and mean in the histogram
    plt.axvline(mean,color='w',linestyle ='dashed',linewidth=2)
    plt.axvline(std,color='w',linestyle ='dashed',linewidth=2)
    plt.axvline(-std,color='w',linestyle ='dashed',linewidth=2)
    
    plt.axvline(mean_SNP,color='r',linestyle ='dashed',linewidth=2)
    plt.axvline(std_SNP,color='r',linestyle ='dashed',linewidth=2)
    plt.axvline(-std_SNP,color='r',linestyle ='dashed',linewidth=2)
    
    plt.show()
    

if __name__ == "__main__":
    run()
```

```python
# import os
import pandas as pd
import numpy as np
import scipy.optimize as spo
import matplotlib.pyplot as plt
%matplotlib inline 

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.ensemble.forest import RandomForestRegressor
from sklearn.metrics import make_scorer
from sklearn.grid_search import GridSearchCV, RandomizedSearchCV
#from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score
from sklearn.cross_validation import ShuffleSplit
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import normalize
import time, os
from datetime import datetime as dt
os.environ['TZ']='UTC'
from datetime import date, datetime

def symbol_to_path(symbol, base_dir="Data"):
    """Return CSV file path given ticker symbol."""
    return os.path.join(base_dir, "{}.csv".format(str(symbol)))

def get_data(symbol, dates):
    df = pd.DataFrame(index=dates)         
    df_temp = pd.read_csv("Data/{}.csv".format(symbol), index_col='Date',parse_dates=True,usecols=['Date','Open', 
                                                                                                   'High', 'Low', 
                                                                                                   'Close', 'Volume', 
                                                                                                   'Adj Close']
                           ,na_values=['nan'])
    df = df.join(df_temp)
    df.fillna(method="ffill", inplace="True") # Forward fill empty trade dates
    df.fillna(method="bfill", inplace="True") # backfill empty trade dates 
    return df


def plot_data(df, title="Stock prices", xlabel="Date", ylabel="Price"):
    ax = df.plot(title=title, figsize=(12,10), fontsize=12)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    plt.show()

    
#function for gridsearch for regression to find the best tuned regressor

def gridsearch(clf, parameters, scorer, X_train, y_train):
    cv = ShuffleSplit(X_train.shape[0], n_iter=10, test_size=0.25, random_state=0)
    n_iter_search = 8
    grid_obj = RandomizedSearchCV(estimator=clf, param_distributions=parameters, n_iter=n_iter_search, scoring=scorer, cv=cv)
    grid_obj = grid_obj.fit(X_train, y_train)
    clf_out = grid_obj.best_estimator_
    return clf_out
    
    
def train_regressor(start_date, end_date, stock_list, days_shift):    
    
    dates = pd.date_range(start_date, end_date) 
    
    #Setting the variables to pick the indices for the training and test datasets, 70%-30% test train split
    
    #print "days between the two training dates"
    date_format = "%Y-%m-%d"
    a = datetime.strptime(start_date, date_format)
    b = datetime.strptime(end_date, date_format)
    delta = b - a
    num_days = delta.days
    #print num_days
    
    #Keeping 70% data for training
    train_num = int(0.7*num_days)
    
    #Accumulating the data for each stock in a dictionary using get_data utility function
    
    symbols = stock_list
    d = {}
    for name in symbols:
        d[name] = get_data(name, dates)
        d[name] = d[name].sort_index(ascending=True, axis=0)
        
        #Following commented out code is used for co-relation and normal distribution analysis. 
        #Please uncomment and run this code to have a look at the pearson coefficients and the distributions 
        '''
        print name
        print "================="+name+"======================="
        print "========Pearson co-relation========"
        print d[name].corr(method='pearson')
        print "Scatter Matrix before transform : "+name
        pd.scatter_matrix(d[name], alpha = 0.3, figsize = (14,8), diagonal = 'kde')
        
        scaler = StandardScaler()

        #Standardizing all the attributes and check the histogram distributions
        scaler.fit(d[name][['Open', 'High', 'Low', 'Close', 'Volume', 'Adj Close']])
        d[name][['Open', 'High', 'Low', 'Close', 'Volume', 'Adj Close']] = scaler.transform(d[name][['Open', 'High', 'Low', 'Close', 'Volume', 'Adj Close']])
        print "Scatter Matrix after transform : "+name
        pd.scatter_matrix(d[name], alpha = 0.3, figsize = (14,8), diagonal = 'kde')
        d[name][['Open', 'High', 'Low', 'Close', 'Volume', 'Adj Close']] = scaler.inverse_transform(d[name][['Open', 'High', 'Low', 'Close', 'Volume', 'Adj Close']])
        '''
 
    #Dropping the three of the four highly co-related attributes - ['High', 'Low', 'Close']
    for key in d:    
        d[key].drop(['High', 'Low', 'Close'], axis=1, inplace=True)
           
    #add new column for Adj Close n days later : "days_shift"
    for key in d:
        d[key]['Adj_Close_req_Days_Later'] = d[key]['Adj Close']
        d[key]['Adj_Close_req_Days_Later'] = d[key]['Adj_Close_req_Days_Later'].shift(-days_shift)
        
        d[key].fillna(method="ffill", inplace="True") # Forward fill empty trade dates
        d[key].fillna(method="bfill", inplace="True") # backfill empty trade dates 
        
    #Intilializing the Linear Regression Regressor
    
    regr = LinearRegression()
    
    #building training models for each stock
    
    lrg_train = {}  
    
    for key in d:
        scaler = StandardScaler()
        X_train = d[key].iloc[0:train_num,:-1]
        scaler.fit(d[key].iloc[0:,:-1])
        print "\n"
        X_train = scaler.transform(X_train)
        y_train = d[key].iloc[0:train_num, -1]
        X_test = d[key].iloc[train_num:num_days,:-1]
        X_test = scaler.transform(X_test)
        y_test = d[key].iloc[train_num:num_days, -1]
        
        
        # Perform gridsearch to find the most optimal hyperparameters for the Linear Regression algorithm
        
        params = {'fit_intercept':[True, False],
                   'normalize' :[True, False],
                   'copy_X': [True, False],
                   'n_jobs':[-1]}
        
        scorer = make_scorer(r2_score)
        
        regr = gridsearch(regr, params, scorer, X_train, y_train)
        
        lrg_train[key] = regr.fit(X_train, y_train)
        
        
        a = key+"_val"
        
        lrg_train[a] = scaler.transform(d[key].iloc[(num_days-1):num_days,:-1])
   
        print "Score on training data for "+key+" :%.2f" % r2_score(y_train, lrg_train[key].predict(X_train))
        print "Score on test data for "+key+" : %.2f" % r2_score(y_test, lrg_train[key].predict(X_test))
        
        X = scaler.transform(d[key].iloc[0:,:-1]) 
        d[key]['Predicted'] = lrg_train[key].predict(X)
        
        #print "The final DataFrame with predicted stock price for :"+key
        #print d[key].head()
        
        #Scatter plot on trained data
       
        ax = d[key].iloc[0:num_days].plot.scatter(x='Adj Close', y='Adj_Close_req_Days_Later')
        d[key].iloc[0:num_days].plot(x='Adj Close',y='Predicted',color='green',ax=ax,label='predicted Adj_Close_req_Days_Later'
                               ,figsize=(18,9),title = key+' Scatter and regression plot ')
        
        #Prediction +/- percentage on the test data
        
        y_diff = ((lrg_train[key].predict(X_test) - y_test)/y_test)*100
        y_diff = y_diff.abs()
        row_cnt = y_diff.shape[0]
        percent_error = round((y_diff.sum(axis=0))/row_cnt,3)
        print "\n"
        print "The prediction on the test dataset "+str(days_shift)+" days after the training date for "+key+" is +/- "+str(percent_error)+"% compared to the actual values"
        
       
    return lrg_train

def query_regressor(train_start, train_end, date_predict, stock_list):    
    
    
    date_format = "%Y-%m-%d"
    a = datetime.strptime(train_end, date_format)
    b = datetime.strptime(date_predict, date_format)
    delta = b - a
    shift = delta.days
    
    regressor = train_regressor(train_start, train_end, stock_list, shift)
    for key in regressor:
        if key in stock_list:
            to_pred = key+"_val"
            predicted = regressor[key].predict(regressor[to_pred])
            print "The predicted stockprice for "+str(key)+" on "+str(date_predict)+" : "+str(predicted)
        else:
            pass

if __name__ == "__main__":
    
    query_regressor('2012-12-01', '2015-12-01','2015-12-11',['AAPL','GOOG', 'AMZN', 'FACEBOOK', 'GE', 'MICROSOFT'])
```

```python
%matplotlib inline 
import pandas as pd
import numpy as np
from statsmodels.tsa.arima_model import ARIMA, ARIMAResults
#from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pylab as plt
#import matplotlib.pyplot as plt
import matplotlib.dates as dates
from sklearn.metrics import r2_score


def symbol_to_path(symbol, base_dir="Data"):
    """Return CSV file path given ticker symbol."""
    return os.path.join(base_dir, "{}.csv".format(str(symbol)))
###['Date', 'Open', 'High', 'Low','Close', 'Volume', 'Adj Close']

def get_data(symbol, dates):
    df = pd.DataFrame(index=dates)  
    df_temp = pd.read_csv("Data/{}.csv".format(symbol), index_col='Date',parse_dates=True,usecols=['Date','Adj Close']
                           ,na_values=['nan'])
    df = df.join(df_temp) #use default how='left'
    #print df.shape
    return df
def plot_data(df, title="Stock prices", xlabel="Date", ylabel="Price"):
    ax = df.plot(title=title, figsize=(12,10), fontsize=12)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    plt.show()

    
def plot_selected(df, columns, start_index, end_index, title="Stock prices"):
    """Plot the desired columns over index values in the given range."""
    # TODO: Your code here
    # Note: DO NOT modify anything else!
    df_new = df.ix[start_index : end_index,columns]
    ax = df_new.plot(title=title,fontsize=15,figsize=(16,10))
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    plt.show()
    
    
def exec_arima(start_date, end_date, stock):    
    
    #prepare the dataframes
    
    dates = pd.date_range(start_date, end_date)
    df = get_data(stock, dates)
    df = df.sort_index(ascending=True, axis=0)
    plot_data(df)
    df.fillna(method="ffill", inplace="True") # Forward fill empty trade dates
    df.fillna(method="bfill", inplace="True") # backfill empty trade dates 
    
    # Get Adj Close One day ahead
    
    df['Adj Close'] = df['Adj Close'].shift(-1)
    
    
    ##DIAGNOSING THE ACF AND PACF PLOTS
    acf = plot_acf(df, lags = 20)
    plt.title("ACF Plot of APPLE")
    acf.show()
    
    pacf = plot_pacf(df, lags = 20)
    plt.title("PACF Plot of APPLE")
    pacf.show()
    
    
    #The ACF graph suggest there is constant auto co-relation. Lets do 1 lag differencing and dig further to confirm this
    #1 lag differencing is not helping much and so decided to do 2 lag differencing

    df_diff = df- df.shift()
    df_diff = df_diff - df_diff.shift()
    df_diff.fillna(method="ffill", inplace="True") # Forward fill empty trade dates
    df_diff.fillna(method="bfill", inplace="True") # backfill empty trade dates 


##DIAGNOSING THE ACF AND PACF PLOTS for DIFF
    acf = plot_acf(df_diff, lags = 20)
    plt.title("ACF Plot of APPLE")
    acf.show()
    
    pacf = plot_pacf(df_diff, lags = 20)
    plt.title("PACF Plot of APPLE")
    pacf.show()
    
    ###It should be an AR model with the shift data as the PACF graph indicates an AR beacuse for an MA model, 
    ###the theoretical PACF does not shut off, but instead tapers toward 0 in some manner.
    
    #Building the ARIMA model
    
    df_train = df['2016-09-01':'2016-11-21']
    
    mod = ARIMA(df_train, order = (0, 2, 1))
    
    results = mod.fit()
    
    predVals = results.predict('2016-10-01', '2016-12-01', typ='levels')
    
    google_pred = pd.concat([df['2014-05-01':'2016-12-01'], predVals], axis = 1)
    
    google_pred = google_pred.rename(columns={'Adj Close':'Original',0:'Predicted'})
    
    google_pred.fillna(0, inplace=True)
    
    ##plot actual vs predicted
    
    plot_selected(google_pred,['Original','Predicted'], '2016-10-01', '2016-12-01',title="Actual Vs. Forecasted Values")
    
    #Calculate R squared on train and test datasets
    
    r2_score_train = r2_score(google_pred['Original']['2016-10-01':'2016-11-21'], google_pred['Predicted']['2016-10-01':'2016-11-21'])
    print "Train score ", r2_score_train
    
    
    r2_score_test = r2_score(google_pred['Original']['2016-11-22':'2016-12-01'], google_pred['Predicted']['2016-11-22':'2016-12-01'])
    print "Test score ", r2_score_test

 
if __name__ == "__main__":
    exec_arima('2011-12-05', '2016-12-02','AAPL')
 
```
